```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.gridspec as gridspec
```

### Figure 4d

```python
fig = plt.figure(dpi=100, figsize=(4,3))
 
gs = gridspec.GridSpec(1, 1, figure=fig)
ax1 = plt.subplot(gs[0, 0])

totalWidth=0.6
labelNums=3
barWidth=totalWidth/labelNums
seriesNums=3 

ax1.bar([x for x in range(seriesNums)], height=[2/5, 2/3, 10/13], label="Weak", width=barWidth,edgecolor='black', color= '#D0CDE2')
ax1.bar([x+barWidth for x in range(seriesNums)], height=[1/3, 3/5, 5/7], label="Moderate", width=barWidth,edgecolor='black', color='#9791C1')
ax1.bar([x+barWidth*2 for x in range(seriesNums)], height=[4/13, 4/7, 20/29], label="Strong", width=barWidth,edgecolor='black', color='#635AA3')

ax1.set_xticks([x+barWidth/2*(labelNums-1) for x in range(seriesNums)], ["GMC","HDB","HIM"], fontsize=10)
ax1.set_ylabel(r"$r^*$", fontsize=12)
ax1.set_yticks(np.arange(0, 1, 0.2))

ax1.set_ylim(0, 0.9)
ax1.tick_params(labelsize=12)
ax1.legend(frameon=False, fontsize=10, loc='lower center', bbox_to_anchor=(0.48, -0.32), ncol=3)

plt.tight_layout(w_pad=3)

mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False

fig.savefig('figure_4d.jpg', bbox_inches = 'tight', dpi=1000)
```
